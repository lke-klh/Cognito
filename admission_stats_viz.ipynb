```python
import pandas as pd
stats = pd.read_csv("Number of applicants by ending status residency MSIM.csv")

stats.head(5)
```

```python
import matplotlib.pyplot as plt

stats['Academic Year'] = stats['Academic Year'].astype(str)

stats_pivot = stats.pivot_table(index='Academic Year', 
                                 columns='Application Status Group', 
                                 values='Application Count (redacted)', aggfunc='sum')

stats_pivot.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Number of Applicants by Application Status and Year")
plt.xlabel("Academic Year")
plt.ylabel("Number of Applicants")
plt.legend(title="Application Status")
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
